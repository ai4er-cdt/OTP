```python
import xarray as xr
```

```python
path = "/mnt/g/My Drive/GTC/solodoch_data"
lats = ["26N", "30S", "55S", "60S"]

ssh_dataset = "ECCO_L4_SSH_05DEG_MONTHLY_V4R4"
stress_dataset = "ECCO_L4_STRESS_05DEG_MONTHLY_V4R4"
obp_dataset = "ECCO_L4_OBP_05DEG_MONTHLY_V4R4"
temp_sal_dataset = "ECCO_L4_TEMP_SALINITY_05DEG_MONTHLY_V4R4"
```

```python
lat = "26N"
```

```python
ssh = xr.open_mfdataset(f"{path}/{lat}/{ssh_dataset}/*.nc",
                        coords="minimal",
                        data_vars="minimal",
                        parallel=True, compat="override")
ssh = ssh["SSH"].to_dataset()

stress = xr.open_mfdataset(f"{path}/{lat}/{stress_dataset}/*.nc",
                           coords="minimal",
                           data_vars="minimal",
                           parallel=True, compat="override")
stress = stress["EXFtaue"].to_dataset()

obp = xr.open_mfdataset(f"{path}/{lat}/{obp_dataset}/*.nc",
                        coords="minimal",
                        data_vars="minimal",
                        parallel=True, compat="override")
obp = obp["OBP"].to_dataset()

ts = xr.open_mfdataset(f"{path}/{lat}/{temp_sal_dataset}/*.nc",
                       coords="minimal",
                       data_vars="minimal",
                       parallel=True, compat="override")
ts = ts[["SALT", "THETA"]].sel(Z=-5.).drop_vars("Z")
```

```python
ts = xr.open_mfdataset(f"{path}/{lat}/{temp_sal_dataset}/*.nc",
                       coords="minimal",
                       data_vars="minimal",
                       parallel=True, compat="override")
ts = ts[["SALT", "THETA"]].sel(Z=-5.).drop_vars("Z")
```

```python
datasets = [ssh, stress, obp, ts]
data = xr.merge(datasets)
var_names = {"SSH": "SSH",
             "EXFtaue": "ZWS",
             "OBP": "OBP",
             "SALT": "SSS",
             "THETA": "SST"}
data = data.rename(var_names)
```
